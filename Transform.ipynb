```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
#%matplotlib inline
#interactive window
%matplotlib qt 
```

```python
filename = 'stop_sign.jpg'
img = mpimg.imread(filename)
plt.imshow(img)

# marks the points on image

plt.plot(310, 298, ".")
plt.plot(276, 412, ".")
plt.plot(845, 108, ".")
plt.plot(926, 189, ".")
```

```python
def warp(img):
    
    img_size = (img.shape[1], img.shape[0])
    
    src = np.float32([
        [310, 298], 
        [276, 412], 
        [845, 108], 
        [926, 189], 
    ])
    
    dest = np.float32([
        [276, 200], 
        [276, 412], 
        [926, 200], 
        [926, 412], 
    ])
    
    M = cv2.getPerspectiveTransform(src, dest)
    
    warped = cv2.warpPerspective(img, M, img_size, flags=cv2.INTER_LINEAR)
    
    return warped    
```

```python
%matplotlib inline

warped_img = warp(img)

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

ax1.set_title("source image")
ax1.imshow(img)

ax2.set_title("destination image")
ax2.imshow(warped_img)
```
